```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Paramètres

np.random.seed(42)
n_users = 800
start_date = datetime(2026, 1, 1)
end_date = datetime(2026, 3, 31)

# Canaux d'acquisition

channels = ['Organic', 'Paid Social', 'Referral', 'Influencer']
channel_weights = [0.35, 0.30, 0.20, 0.15]

# Pays

countries = ['France', 'Germany', 'Spain', 'Italy', 'Belgium']
country_weights = [0.45, 0.25, 0.15, 0.10, 0.05]

# Catégories produits

categories = ['Smartphones', 'Laptops', 'Tablets', 'Audio', 'Gaming']

# Générer les users

users = pd.DataFrame({
'user_id': range(1, n_users + 1),
'signup_date': [start_date + timedelta(days=random.randint(0, 89)) for _ in range(n_users)],
'acquisition_channel': np.random.choice(channels, n_users, p=channel_weights),
'country': np.random.choice(countries, n_users, p=country_weights)
})

print(users.head())
print(f"Total users: {len(users)}")
```

```python
# Générer les orders
orders_list = []

for _, user in users.iterrows():
    # Probabilité d'achat selon le canal
    if user['acquisition_channel'] == 'Paid Social':
        purchase_prob = 0.55
    elif user['acquisition_channel'] == 'Influencer':
        purchase_prob = 0.50
    elif user['acquisition_channel'] == 'Referral':
        purchase_prob = 0.65
    else:  # Organic
        purchase_prob = 0.45

    # Nombre de commandes
    if random.random() < purchase_prob:
        n_orders = random.randint(1, 4)
        for _ in range(n_orders):
            order_date = user['signup_date'] + timedelta(days=random.randint(1, 89))
            if order_date <= end_date:
                orders_list.append({
                    'order_id': len(orders_list) + 1,
                    'user_id': user['user_id'],
                    'order_date': order_date,
                    'order_value': round(random.uniform(50, 800), 2),
                    'product_category': random.choice(categories)
                })

orders = pd.DataFrame(orders_list)
print(orders.head())
print(f"Total orders: {len(orders)}")
print(f"Unique buyers: {orders['user_id'].nunique()}")
```

```python
sessions_list = []

for _, user in users.iterrows():
    # Nombre de sessions selon si l'user a acheté ou pas
    has_ordered = user['user_id'] in orders['user_id'].values

    if has_ordered:
        n_sessions = random.randint(3, 15)
    else:
        n_sessions = random.randint(1, 5)

    for _ in range(n_sessions):
        session_date = user['signup_date'] + timedelta(days=random.randint(0, 89))
        if session_date <= end_date:
            sessions_list.append({
                'session_id': len(sessions_list) + 1,
                'user_id': user['user_id'],
                'session_date': session_date
            })

sessions = pd.DataFrame(sessions_list)
print(sessions.head())
print(f"Total sessions: {len(sessions)}")
print(f"Unique users with sessions: {sessions['user_id'].nunique()}")
```

```python
users.to_csv('reloop_users.csv', index=False)
orders.to_csv('reloop_orders.csv', index=False)
sessions.to_csv('reloop_sessions.csv', index=False)

print("✅ Les 3 fichiers CSV sont prêts !")
```

```python
import sqlite3

# Créer la base de données
conn = sqlite3.connect('reloop.db')

# Charger les 3 tables
users.to_sql('users', conn, if_exists='replace', index=False)
orders.to_sql('orders', conn, if_exists='replace', index=False)
sessions.to_sql('sessions', conn, if_exists='replace', index=False)

print("✅ Base de données créée avec 3 tables !")

# Vérification
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
print("Tables disponibles :", cursor.fetchall())
```

```python
query = """
SELECT
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT o.user_id) AS activated_users,
    ROUND(COUNT(DISTINCT o.user_id) * 100.0 / COUNT(DISTINCT u.user_id), 1) AS activation_rate
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    u.acquisition_channel,
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT o.user_id) AS activated_users,
    ROUND(COUNT(DISTINCT o.user_id) * 100.0 / COUNT(DISTINCT u.user_id), 1) AS activation_rate
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
GROUP BY u.acquisition_channel
ORDER BY activation_rate DESC
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    COUNT(DISTINCT first_orders.user_id) AS total_activated_users,
    COUNT(DISTINCT CASE
        WHEN JULIANDAY(second_orders.order_date) - JULIANDAY(first_orders.first_order_date) <= 7
        THEN second_orders.user_id
    END) AS retained_d7,
    COUNT(DISTINCT CASE
        WHEN JULIANDAY(second_orders.order_date) - JULIANDAY(first_orders.first_order_date) <= 30
        THEN second_orders.user_id
    END) AS retained_d30,
    ROUND(COUNT(DISTINCT CASE
        WHEN JULIANDAY(second_orders.order_date) - JULIANDAY(first_orders.first_order_date) <= 7
        THEN second_orders.user_id
    END) * 100.0 / COUNT(DISTINCT first_orders.user_id), 1) AS retention_d7_rate,
    ROUND(COUNT(DISTINCT CASE
        WHEN JULIANDAY(second_orders.order_date) - JULIANDAY(first_orders.first_order_date) <= 30
        THEN second_orders.user_id
    END) * 100.0 / COUNT(DISTINCT first_orders.user_id), 1) AS retention_d30_rate
FROM (
    SELECT user_id, MIN(order_date) AS first_order_date
    FROM orders
    GROUP BY user_id
) AS first_orders
LEFT JOIN orders second_orders
    ON first_orders.user_id = second_orders.user_id
    AND second_orders.order_date > first_orders.first_order_date
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    u.acquisition_channel,
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT o.order_id) AS total_orders,
    ROUND(SUM(o.order_value), 2) AS total_revenue,
    ROUND(SUM(o.order_value) / COUNT(DISTINCT u.user_id), 2) AS arpu,
    ROUND(SUM(o.order_value) / COUNT(DISTINCT o.order_id), 2) AS avg_order_value
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
GROUP BY u.acquisition_channel
ORDER BY arpu DESC
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT CASE
        WHEN last_session.last_session_date < DATE('2026-03-01')
        THEN u.user_id
    END) AS churned_users,
    COUNT(DISTINCT CASE
        WHEN last_session.last_session_date >= DATE('2026-03-01')
        THEN u.user_id
    END) AS active_users,
    ROUND(COUNT(DISTINCT CASE
        WHEN last_session.last_session_date < DATE('2026-03-01')
        THEN u.user_id
    END) * 100.0 / COUNT(DISTINCT u.user_id), 1) AS churn_rate
FROM users u
LEFT JOIN (
    SELECT user_id, MAX(session_date) AS last_session_date
    FROM sessions
    GROUP BY user_id
) AS last_session ON u.user_id = last_session.user_id
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
# Taux d'activation global
query1 = """
SELECT
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT o.user_id) AS activated_users,
    ROUND(COUNT(DISTINCT o.user_id) * 100.0 / COUNT(DISTINCT u.user_id), 1) AS activation_rate
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
"""

# Activation par canal
query2 = """
SELECT
    u.acquisition_channel,
    COUNT(DISTINCT u.user_id) AS total_users,
    COUNT(DISTINCT o.user_id) AS activated_users,
    ROUND(COUNT(DISTINCT o.user_id) * 100.0 / COUNT(DISTINCT u.user_id), 1) AS activation_rate
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
GROUP BY u.acquisition_channel
ORDER BY activation_rate DESC
"""

# ARPU par canal
query3 = """
SELECT
    u.acquisition_channel,
    COUNT(DISTINCT u.user_id) AS total_users,
    ROUND(SUM(o.order_value) / COUNT(DISTINCT u.user_id), 2) AS arpu,
    ROUND(SUM(o.order_value) / COUNT(DISTINCT o.order_id), 2) AS avg_order_value
FROM users u
LEFT JOIN orders o ON u.user_id = o.user_id
GROUP BY u.acquisition_channel
ORDER BY arpu DESC
"""

# Sauvegarder
pd.read_sql_query(query1, conn).to_csv('reloop_activation.csv', index=False)
pd.read_sql_query(query2, conn).to_csv('reloop_activation_by_channel.csv', index=False)
pd.read_sql_query(query3, conn).to_csv('reloop_arpu_by_channel.csv', index=False)

print("✅ Résultats exportés !")
```
